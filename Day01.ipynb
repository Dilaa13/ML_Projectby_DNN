```python
import pandas as pd

# Load dataset
train_df = pd.read_csv('/home/charith-dhananjaya/Desktop/ML Project/train.csv')

# Check for duplicate rows
print("\nDuplicate Rows in Train Data:", train_df.duplicated().sum())

# Drop duplicate rows if necessary
train_df = train_df.drop_duplicates()

# Separate numerical and categorical columns
numerical_features = train_df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = train_df.select_dtypes(include=['object']).columns

# Handling missing values:
# - Fill numerical columns with their mean
# - Fill categorical columns with the most frequent (mode) value

train_df[numerical_features] = train_df[numerical_features].fillna(train_df[numerical_features].mean())

# Fix: Assign categorical filling separately to avoid Pandas 3.0 warning
for col in categorical_features:
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])

# Confirm missing values are handled
print("\nMissing values after handling:")
print(train_df.isnull().sum())
```

- There are no duplicate rows in traing data

```python

# Load dataset
train_df = pd.read_csv('/home/charith-dhananjaya/Desktop/ML Project/train.csv')

# Separate numerical and categorical columns
numerical_features = train_df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = train_df.select_dtypes(include=['object']).columns

# 1 Histograms with KDE for Numerical Features
for col in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(train_df[col], kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# 2 Boxplots for Outlier Detection in Numerical Features
for col in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=train_df[col], color="red")
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

# 3 Count Plots for Categorical Features
for col in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(y=train_df[col], palette="viridis")
    plt.title(f"Count Plot of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.show()

# 4 Correlation Heatmap (for numerical features only)
plt.figure(figsize=(12, 6))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
train_df = pd.read_csv('/home/charith-dhananjaya/Desktop/ML Project/train.csv')

# Set up a larger figure size for better clarity
plt.figure(figsize=(18, 12))

# Generate an enhanced correlation heatmap with larger annotations
sns.heatmap(train_df.corr(numeric_only=True), 
            annot=True,           # Show correlation values
            cmap="coolwarm",       # Red = Positive, Blue = Negative
            fmt=".2f",             # Limit decimals to 2 places
            linewidths=1,          # Increase gridline thickness
            square=True,           # Keep cells square for readability
            annot_kws={"size": 14}, # Increase text size inside cells
            cbar_kws={'shrink': 0.5})  # Adjust colorbar size

# Increase title size for better visibility
plt.title("Enhanced & Enlarged Feature Correlation Heatmap", fontsize=20)

plt.show()
```
